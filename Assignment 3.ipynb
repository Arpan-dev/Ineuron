```python
lst=[]
n=int(input("Enter a numbers of element: "))  
for i in range(0, n):
    ele = int(input())
 
    lst.append(ele) # adding the element
     


def myfilter(*args):   #filter out even or odd number from a list
    num_even_list=[]
    num_odd_list=[]
    
    for i in lst:
        if i%2==0:
            num_even_list.append(i)
        else:
            num_odd_list.append(i)
            
    return num_even_list,num_odd_list

output=myfilter(n)

print("List is: ",lst)
print("List of Even numbers is: ",output[0])
print("List of Odd numbers is: ",output[1])
```
